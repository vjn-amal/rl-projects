```python
alpha=0.5
rd=0
qval=0
selected_index=0
training=5
```

```python
import pandas as pd

dataframe=pd.read_csv("conversation.csv",sep=';',dtype={'qval':float})

dataframe
```

```python
import random

def random_answer(conv):
    r_frame=dataframe.loc[dataframe['conv']==conv]
    r_choice=random.choice(r_frame.index)
    selected_df=r_frame.iloc[r_choice]
    return selected_df
```

```python
def chat():
    print("Ask your question \n")
    q=input()
    print(random_answer(1).answers)
    fb=input("Did I answer your question ? Please give your feedback (Y/N)")
    print("Thankyou !")
    return fb
```

```python
def get_reward(fb):
    if(fb.lower()=='y'):
        rd=1
    else:
        rd=-1
    return rd
```

```python
def update_qval(qval,rd,alpha):
#     print('qval',qval,'rd',rd,'alpha',alpha)
    new_qval=qval+(alpha*rd)-(alpha*qval)
#     print(new_qval)
    return new_qval
```

```python
def episode():
    q=input("BOT: Ask your question")
    selected_df=random_answer(1)
    print("BOT:",selected_df.answers)
    fb=input("BOT: Did I answer your question ? Please give your feedback (Y/N)")
    print("BOT: Thankyou !")

    rd=get_reward(fb)
    new_qval=update_qval(selected_df.qval,rd,alpha)
    return dataframe.set_value(selected_df.slno,'qval',new_qval)
```

```python
for i in range(training):
    dataframe=episode()
dataframe
```

```python
dataframe.to_csv('qtable.csv', sep=';',index=False)
```
